```python
df = df.drop(columns=['ORDERNUMBER','ORDERDATE','ADDRESSLINE1','ADDRESSLINE2','CONTACTLASTNAME','CONTACTFIRSTNAME','CUSTOMERNAME','PHONE'])
```

```python
df.drop(columns=['STATE','POSTALCODE','TERRITORY','QTR_ID','MONTH_ID'],inplace=True)
```

```python
df
```

```python
df.duplicated().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
en = LabelEncoder()
```

```python
df['STATUS'] = en.fit_transform(df['STATUS'])
df['PRODUCTLINE'] = en.fit_transform(df['PRODUCTLINE'])
df['PRODUCTCODE'] = en.fit_transform(df['PRODUCTCODE'])
df['CITY'] = en.fit_transform(df['CITY'])
df['COUNTRY'] = en.fit_transform(df['COUNTRY'])
df['DEALSIZE'] = en.fit_transform(df['DEALSIZE'])
```

```python
df.head()
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,21):
    km = KMeans(n_clusters=i)
    km.fit_predict(df)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(wcss)
```

```python
km = KMeans(n_clusters=4)
pred = km.fit_predict(df)
```

```python
df['label'] = pred
```

```python
# K-Means Clustering: This is a method used to group data points into clusters (or groups) based on their similarities. You start by picking a number of clusters (K), and the algorithm divides the data into K clusters so that points within each cluster are close to each other. The goal is to minimize the distance between data points and the center (mean) of their cluster, hence "K-Means."

# Hierarchical Clustering: This is another way to group data into clusters, but it builds a hierarchy (a tree) of clusters. It starts with each data point as its own cluster and then gradually merges the closest clusters together, one at a time, until everything is in one big cluster. This method helps show how data points are grouped at different levels.

# Elbow Method: The elbow method helps find the best number of clusters (K) to use in K-means. It involves running K-means for different values of K (e.g., from 1 to 10) and plotting the "Within-Cluster Sum of Squares" (WCSS), which measures how tight the clusters are. As K increases, WCSS decreases, but at a certain point (the "elbow"), the improvement slows down. The "elbow" point suggests an ideal number of clusters.
```
